Occupations
Step 1. Import the necessary libraries
Step 2. Import the dataset from this address.
Step 3. Assign it to a variable called users
Step 4. Discover what is the mean age per occupation
Step 5. Discover the Male ratio per occupation and sort it from the most to the least
Step 6. For each occupation, calculate the minimum and maximum ages
Step 7. For each combination of occupation and sex, calculate the mean age
Step 8. For each occupation present the percentage of women and men

```python
# Step 1. Import the necessary libraries
import pandas as pd

# Step 2. Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|')

# Step 3. Assign it to a variable called users
# This step is done in the previous step

# Step 4. Discover what is the mean age per occupation
mean_age = users.groupby('occupation')['age'].mean()
display(mean_age)
```

```python
# Step 5. Discover the Male ratio per occupation and sort it from the most to the least
male_ratio = users[users['gender'] == 'M'].groupby('occupation')['gender'].count() / users.groupby('occupation')['gender'].count()
male_ratio = male_ratio.sort_values(ascending=False)
display(male_ratio)
```

```python
# Step 6. For each occupation, calculate the minimum and maximum ages
min_max_age = users.groupby('occupation')['age'].agg(['min', 'max'])
display(min_max_age)
```

```python
# Step 7. For each combination of occupation and sex, calculate the mean age
mean_age_sex = users.groupby(['occupation', 'gender'])['age'].mean()
display(mean_age_sex)
```

```python
# Step 8. For each occupation present the percentage of women and men
gender_ocup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
occup_count = users.groupby(['occupation']).agg('count')
gender_ocup_ratio = gender_ocup.div(occup_count, level = "occupation") * 100
display(gender_ocup_ratio)
```

Wind Statistics 
The data have been modified to contain some missing values, identified by NaN.
Using pandas should make this exercise easier, in particular for the bonus question.
You should be able to perform all of these operations without using a for loop or 
other looping construct.
Step 1. Import the necessary libraries
Step 2. Import the dataset from the attached file wind.txt
Step 3. Assign it to a variable called data and replace the first 3 columns by a proper 
datetime index.
Step 4. Year 2061? Do we really have data from this year? Create a function to fix it 
and apply it.
Step 5. Set the right dates as the index. Pay attention at the data type, it should be 
datetime64[ns].
Step 6. Compute how many values are missing for each location over the entire 
record.They should be ignored in all calculations below.
Step 7. Compute how many non-missing values there are in total.
Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and 
all the times.
A single number for the entire dataset.
Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean 
windspeeds and standard deviations of the windspeeds at each location over all the 
days
A different set of numbers for each location.
Step 10. Create a DataFrame called day_stats and calculate the min, max and mean 
windspeed and standard deviations of the windspeeds across all the locations at each 
day.
A different set of numbers for each day.
Step 11. Find the average windspeed in January for each location.
Treat January 1961 and January 1962 both as January.
Step 12. Downsample the record to a yearly frequency for each location.
Step 13. Downsample the record to a monthly frequency for each location.
Step 14. Downsample the record to a weekly frequency for each location.
Step 15. Calculate the min, max and mean windspeeds and standard deviations of the

January 2 1961) for the first 52 weeks.

```python
# Step 1: Import the necessary libraries
import pandas as pd
import numpy as np

# Step 2: Import the dataset
file_path = 'wind.txt'
data = pd.read_csv(file_path, sep='\s+', parse_dates=[[0, 1, 2]])

# Step 3: Replace the first 3 columns by a proper datetime index
data.columns = ['Date', 'RPT', 'VAL', 'ROS', 'KIL', 'SHA', 'BIR', 'DUB', 'CLA', 'MUL', 'CLO', 'BEL', 'MAL']

# Print the output after step 3
data.head()
```

```python
# Step 4: Fix the year 2061 issue
def fix_year(date):
    year = date.year - 100 if date.year > 1989 else date.year
    return pd.Timestamp(year, date.month, date.day)

data['Date'] = data['Date'].apply(fix_year)

# Print the output after step 4
data.head()
```

```python
# Step 5: Set the right dates as the index
data.set_index('Date', inplace=True)
data.index = pd.to_datetime(data.index)

# Print the output after step 5
data.head()
```

```python
# Step 6: Compute how many values are missing for each location
missing_values = data.isnull().sum()

print(missing_values)
```

```python
# Step 7: Compute how many non-missing values there are in total
non_missing_values = data.count().sum()

print(non_missing_values)
```

```python
# Step 8: Calculate the mean windspeeds of the windspeeds over all the locations and all the times
mean_windspeed_all = data.mean().mean()

print(mean_windspeed_all)
```

```python
# Step 9: Create loc_stats DataFrame
loc_stats = data.describe()

print(loc_stats.head())
```

```python
# Step 10: Create day_stats DataFrame
day_stats = data.agg(['min', 'max', 'mean', 'std'], axis=1)

print (day_stats.head())
```

```python
# Step 11: Find the average windspeed in January for each location
january_data = data[data.index.month == 1]
january_means = january_data.mean()

print(january_means)
```

```python
# Step 12: Downsample to a yearly frequency for each location
yearly_data = data.resample('Y').mean()

print(yearly_data.head())
```

```python
# Step 13: Downsample to a monthly frequency for each location
monthly_data = data.resample('M').mean()

print(monthly_data.head())
```

```python
# Step 14: Downsample to a weekly frequency for each location
weekly_data = data.resample('W').mean()

print(weekly_data.head())
```

```python
# Step 15: Calculate the min, max and mean windspeeds and standard deviations for each week (for the first 52 weeks)
weekly_stats_first_52 = data.resample('W').agg(['min', 'max', 'mean', 'std']).iloc[1:53]

print(weekly_stats_first_52.head())
```

Question 6
Create a line plot showing the number of marriages and divorces per capita in the 
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import matplotlib.pyplot as plt
import pandas as pd  

# Import the CSV file into a DataFrame
file_path = 'us-marriages-divorces-1867-2014.csv'
us_data = pd.read_csv(file_path)

# Create the line plot
plt.figure(figsize=(12, 6))
plt.plot(us_data['Year'], us_data['Marriages_per_1000'], label='Marriages Per Capita', color='b')
plt.plot(us_data['Year'], us_data['Divorces_per_1000'], label='Divorces Per Capita', color='r')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Per 1000 People')
plt.title('Number of Marriages and Divorces Per Capita in the U.S. (1867-2014)')

# Show legend
plt.legend()

# Display the plot
plt.show()
```

Question 7
Create a vertical bar chart comparing the number of marriages and divorces per 
capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
# Filter the data for the years 1900, 1950, and 2000
filtered_data = us_data[us_data['Year'].isin([1900, 1950, 2000])]

# Create the vertical bar chart
plt.figure(figsize=(10, 6))

barWidth = 0.4

# Set position of bar on X axis
r1 = np.arange(len(filtered_data['Year']))
r2 = [x + barWidth for x in r1]

plt.bar(r1, filtered_data['Marriages_per_1000'], color ='b', width = barWidth, edgecolor ='grey', label ='Marriages Per Capita')
plt.bar(r2, filtered_data['Divorces_per_1000'], color ='r', width = barWidth, edgecolor ='grey', label ='Divorces Per Capita')

# Add labels to the bar
plt.xlabel('Years', fontweight ='bold')
plt.ylabel('Per 1000 People')
plt.xticks([r + barWidth / 2 for r in range(len(filtered_data['Year']))], filtered_data['Year'])

plt.title('Number of Marriages and Divorces Per Capita in the U.S. for the Years 1900, 1950, and 2000')
plt.legend()

# Show plot
plt.show()
```

Question 9
Create a pie chart showing the fraction of all Roman Emperors that were 
assassinated.
Make sure that the pie chart is an even circle, labels the categories, and shows the 
percentage breakdown of the categories.

```python
# Import the CSV file into a DataFrame
emperor_file_path = 'roman-emperor-reigns.csv'
emperor_data = pd.read_csv(emperor_file_path)

# Count the occurrences of each 'Cause_of_Death'
cause_of_death_counts = emperor_data['Cause_of_Death'].value_counts()

# Calculate the fraction of emperors that were assassinated
assassinated_fraction = cause_of_death_counts['Assassinated'] / cause_of_death_counts.sum()

# Data for the pie chart
labels = ['Assassinated', 'Others']
sizes = [assassinated_fraction, 1 - assassinated_fraction]
colors = ['red', 'blue']
explode = (0.1, 0)  # explode the 1st slice (Assassinated)

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Fraction of Roman Emperors Assassinated')

# Show plot
plt.show()
```
