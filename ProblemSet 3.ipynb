Occupations
Step 1. Import the necessary libraries
Step 2. Import the dataset from this address.
Step 3. Assign it to a variable called users
Step 4. Discover what is the mean age per occupation
Step 5. Discover the Male ratio per occupation and sort it from the most to the least
Step 6. For each occupation, calculate the minimum and maximum ages
Step 7. For each combination of occupation and sex, calculate the mean age
Step 8. For each occupation present the percentage of women and men

```python
# Step 1. Import the necessary libraries
import pandas as pd

# Step 2. Import the dataset
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|')

# Step 3. Assign it to a variable called users
# This step is done in the previous step

# Step 4. Discover what is the mean age per occupation
mean_age = users.groupby('occupation')['age'].mean()
display(mean_age)
```

```python
# Step 5. Discover the Male ratio per occupation and sort it from the most to the least
male_ratio = users[users['gender'] == 'M'].groupby('occupation')['gender'].count() / users.groupby('occupation')['gender'].count()
male_ratio = male_ratio.sort_values(ascending=False)
display(male_ratio)
```

```python
# Step 6. For each occupation, calculate the minimum and maximum ages
min_max_age = users.groupby('occupation')['age'].agg(['min', 'max'])
display(min_max_age)
```

```python
# Step 7. For each combination of occupation and sex, calculate the mean age
mean_age_sex = users.groupby(['occupation', 'gender'])['age'].mean()
display(mean_age_sex)
```

```python
# Step 8. For each occupation present the percentage of women and men
gender_ocup = users.groupby(['occupation', 'gender']).agg({'gender': 'count'})
occup_count = users.groupby(['occupation']).agg('count')
gender_ocup_ratio = gender_ocup.div(occup_count, level = "occupation") * 100
display(gender_ocup_ratio)
```

Euro Teams
Step 1. Import the necessary libraries
Step 2. Import the dataset from this address
Step 3. Assign it to a variable called euro12
Step 4. Select only the Goal column
Step 5. How many team participated in the Euro2012?
Step 6. What is the number of columns in the dataset?
Step 7. View only the columns Team, Yellow Cards and Red Cards and assign them 
to a dataframe called discipline
Step 8. Sort the teams by Red Cards, then to Yellow Cards
Step 9. Calculate the mean Yellow Cards given per Team
Step 10. Filter teams that scored more than 6 goalsStep 11. Select the teams that start 
with G
Step 12. Select the first 7 columns
Step 13. Select all columns except the last 3
Step 14. Present only the Shooting Accuracy from England, Italy and Russia

```python
# Step 1. Import the necessary libraries
import pandas as pd

# Step 2. Import the dataset from the given address
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"

# Step 3. Assign it to a variable called euro12
euro12 = pd.read_csv(url)

# Step 4. Select only the Goal column
goals = euro12['Goals']
display(euro12['Goals'])

# Step 5: Display number of teams that participated in the Euro2012
num_teams = euro12['Team'].nunique()
display(num_teams)

# Step 6: Display number of columns in the dataset
num_columns = euro12.shape[1]
display(num_columns)

# Step 7: Display 'Team', 'Yellow Cards', and 'Red Cards' columns
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
display(discipline)

# Step 8: Sort and display teams by Red Cards, then by Yellow Cards
sorted_discipline = discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=[False, False])
display(sorted_discipline)

# Step 9: Display mean Yellow Cards given per Team
mean_yellow_cards = discipline['Yellow Cards'].mean()
display(mean_yellow_cards)

# Step 10: Display teams that scored more than 6 goals
teams_more_than_6_goals = euro12[euro12['Goals'] > 6]
display(teams_more_than_6_goals)

# Step 11: Display teams that start with 'G'
teams_start_with_G = euro12[euro12['Team'].str.startswith('G')]
display(teams_start_with_G)

# Step 12: Display the first 7 columns
first_7_columns = euro12.iloc[:, :7]
display(first_7_columns)

# Step 13: Display all columns except the last 3
all_except_last_3 = euro12.iloc[:, :-3]
display(all_except_last_3)

# Step 14: Display Shooting Accuracy from England, Italy, and Russia
shooting_accuracy = euro12[euro12['Team'].isin(['England', 'Italy', 'Russia'])][['Team', 'Shooting Accuracy']]
display(shooting_accuracy)
```

Step 2. Create 3 differents Series, each of length 100, as follows:
• The first a random number from 1 to 4
• The second a random number from 1 to 3
• The third a random number from 10,000 to 30,000

```python
# Step 1: Import the necessary libraries
import pandas as pd
import numpy as np

# Step 2: Create 3 different Series, each of length 100
s1 = pd.Series(np.random.randint(1, 5, 100))
s2 = pd.Series(np.random.randint(1, 4, 100))
s3 = pd.Series(np.random.randint(10000, 30001, 100))

# Step 3: Create a DataFrame by joining the Series by column
df = pd.concat([s1, s2, s3], axis=1)

# Step 4: Change the name of the columns to bedrs, bathrs, price_sqr_meter
df.columns = ['bedrs', 'bathrs', 'price_sqr_meter']

# Print the first 5 rows of the DataFrame
print("First 5 rows of the DataFrame:")
print(df.head())
```

```python
# Step 5: Create a one column DataFrame with the values of the 3 Series and assign it to 'bigcolumn'
bigcolumn = pd.concat([s1, s2, s3], axis=0).reset_index(drop=True)
bigcolumn_df = pd.DataFrame(bigcolumn, columns=['bigcolumn'])

# Print the first 5 rows of the bigcolumn DataFrame
print("First 5 rows of the 'bigcolumn' DataFrame:")
print(bigcolumn_df.head())
```

```python
# Step 6: Check if the index goes only until 99
index_check = bigcolumn.index[-1]

# Print the last index of bigcolumn
print(f"The last index of 'bigcolumn': {index_check}")
```

```python
# Step 7: Reindex the DataFrame so it goes from 0 to 299
bigcolumn_df.reset_index(drop=True, inplace=True)

# Print the last 5 rows of the reindexed DataFrame
print("\nLast 5 rows of the reindexed 'bigcolumn' DataFrame:")
print(bigcolumn_df.tail())
```
